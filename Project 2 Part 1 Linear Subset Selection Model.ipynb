# Diamond Price Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
file_path = '/Users/ajayiyer/Desktop/DAT 402/Project 2/diamonds.csv'

df = pd.read_csv(file_path)
df
```

```python
df.drop(df.columns[[0]], axis=1, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.rename(columns={'depth': 'depth_prcnt', 'table': 'table_prcnt', 'x': 'length', 'y': 'width', 'z': 'depth'}, inplace=True)
df
```

```python
le = LabelEncoder()

le.fit(df.cut.drop_duplicates())
df.cut = le.transform(df.cut)

le.fit(df.color.drop_duplicates())
df.color = le.transform(df.color)

le.fit(df.clarity.drop_duplicates())
df.clarity = le.transform(df.clarity)

df
```

```python
column_to_move = df.pop("price")

df.insert(9, "price", column_to_move)
df
```
